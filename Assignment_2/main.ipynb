```markdown
# üõ°Ô∏è Assignment 2 ‚Äì Cybersecurity Visualization Notebook

This notebook explores types of cyberattacks and their common targets using three types of visualizations:

1. Bar chart ‚Äì Incident count by attack type  
2. Network graph ‚Äì Attack vectors and targets  
3. Matrix ‚Äì Tactics vs Techniques (MITRE-inspired)
```

#### üìå Imports

````python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
````

## üìä Bar Chart ‚Äì Incident Count by Attack Type

#### üìå Bar Chart

````python
attack_types = [
    "Phishing", "Ransomware", "DDoS", "SQL Injection", "Credential Stuffing",
    "Insider Threat", "Malware", "Zero-Day Exploit", "Social Engineering"
]
incident_counts = [480, 420, 360, 220, 190, 150, 510, 95, 370]
incident_counts, attack_types = zip(*sorted(zip(incident_counts, attack_types), reverse=True))

plt.figure(figsize=(10, 6))
bars = plt.barh(attack_types, incident_counts, color='tomato')
plt.xlabel("Number of Incidents (simulated)")
plt.title("Cyber Attacks by Type")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
````

## üåê Network Graph ‚Äì Attack Vectors and Targets

#### Network Graph

````python
G = nx.Graph()
edges = [
    ("Phishing", "Individuals"),
    ("Ransomware", "Healthcare"),
    ("Ransomware", "SMBs"),
    ("DDoS", "Banks"),
    ("DDoS", "Govt Sites"),
    ("SQL Injection", "Web Apps"),
    ("Credential Stuffing", "E-commerce"),
    ("Insider Threat", "Internal DB"),
    ("Zero-Day Exploit", "Govt Agencies"),
    ("Social Engineering", "Employees")
]
G.add_edges_from(edges)
pos = nx.spring_layout(G, seed=42)

plt.figure(figsize=(10, 8))
nx.draw_networkx_nodes(G, pos, node_size=1200, node_color='lightblue')
nx.draw_networkx_edges(G, pos, edge_color='gray')
nx.draw_networkx_labels(G, pos, font_size=10)
plt.title("Cyber Attack Vectors and Their Targets")
plt.axis("off")
plt.tight_layout()
plt.show()
````

## üß± Matrix ‚Äì Tactics vs Techniques (MITRE-Inspired)

#### Matrix Grid

````python
tactics = ["Initial Access", "Execution", "Persistence", "Exfiltration"]
techniques = ["Phishing", "Malware", "Credential Dumping", "C2 Traffic", "Lateral Movement"]

np.random.seed(0)
data = np.random.randint(0, 10, size=(len(techniques), len(tactics)))

fig, ax = plt.subplots(figsize=(8, 5))
im = ax.imshow(data, cmap="Reds")

ax.set_xticks(np.arange(len(tactics)))
ax.set_yticks(np.arange(len(techniques)))
ax.set_xticklabels(tactics)
ax.set_yticklabels(techniques)

for i in range(len(techniques)):
    for j in range(len(tactics)):
        ax.text(j, i, data[i, j], ha="center", va="center", color="black")

plt.title("Simplified ATT&CK Matrix (simulated prevalence)")
plt.colorbar(im, ax=ax)
plt.tight_layout()
plt.show()
````
